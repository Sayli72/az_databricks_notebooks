#### Schema should be specified while reading streaming data

```
# Define the schema using StructType
streaming_schema_test = StructType([
    StructField("timestamp", DoubleType(), True),
    StructField("user_id", StringType(), True),
    StructField("event_type", StringType(), True),
    StructField("device_type", StringType(), True),
    StructField("session_id", StringType(), True),
    StructField("EventProcessedUtcTime", TimestampType(), True),
    StructField("PartitionId", LongType(), True),
    StructField("EventEnqueuedUtcTime", TimestampType(), True) 
])
```

```
streaming_df_test = spark.readStream \
                    .format("parquet") \
                    .option("mode", "PERMISSIVE") \
                    .option("header","true")\
                    .schema(streaming_schema_test)\
                    .load(streaming_path)
```

```
streaming_df_test.display()
```

```
Test_df = streaming_df_test.groupBy().count()
```

```
Test_df.display()
```

```
streaming_df_test_2 = spark.readStream \
                    .format("parquet") \
                    .option("columnNameOfCorruptRecord", "CorruptRecord")\
                    .option("header","true")\
                    .schema(streaming_schema_test)\
                    .load(streaming_path)
```
